```python
import pandas as pd
import numpy as np
import re
data = pd.read_excel('instagram_hashtag_data_f.xlsx')
df = data.drop('id',axis=1)
df
```

```python
df['word'] = df['word'].apply(lambda x : re.sub("<[a-zA-Z0-9+]+>|,",'', str(x))) 
df['word'] = df['word'].apply(lambda x: x.replace('\\t',''))
#df['word'] = df['word'].apply(lambda x : re.sub("[^가-힣a-zA-Z0-9, ]+",'', str(x))) 
#df['word'] = df['word'].apply(lambda x : re.sub("[가-힣]t",'', str(x))) 
```

```python
x = df.word
y = df.type
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import SimpleRNN, Embedding, Dense
from tensorflow.keras.models import Sequential

token = Tokenizer()
token.fit_on_texts(df['word'])


train_X, test_X, train_y, test_y = train_test_split(x,y,test_size=0.2, random_state=1234)

X_train_encode = token.texts_to_sequences(train_X)
word_to_index = token.word_index
pad_x = pad_sequences(X_train_encode)
test_X_encode = token.texts_to_sequences(test_X)
X_test_pad = pad_sequences(test_X_encode)
vocab_size = len(word_to_index) + 1
```

```python
model = Sequential()
model.add(Embedding(vocab_size, 32)) # 임베딩 벡터의 차원은 32
model.add(SimpleRNN(32)) # RNN 셀의 hidden_size는 32
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
history = model.fit(pad_x, train_y, epochs=20, batch_size=64, validation_split=0.2)
```

```python
print("\n 테스트 정확도: %.4f" % (model.evaluate(X_test_pad, test_y)[1]))
model.predict(X_test_pad)
```
